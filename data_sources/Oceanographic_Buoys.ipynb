<h3>AUXILIARY FUNCTIONS</h3>

```python
def params_concatenation(params, values):
    """
    Concatenates the params and their values in an unique string to later be attached to the generic API ENDPOINTS
    """
    params_concatenation = ''
    for param, value in zip(params,values):
        if params.index(param) != len(PARAMS)-1:
            params_concatenation = params_concatenation + param+'='+value+'&'
        else:
            params_concatenation = params_concatenation + param+'='+value
    return params_concatenation;
```

<H3>DATA-SOURCES: OCEANOGRAPHIC BUOYS</H3>

When looking for specific platforms it is handy to remember the PLATFORM-TYPE ENDPOINT (see [Quick start](https://github.com/pazrg/SOCIB_API/blob/master/tips/quick_start.ipynb). According to this, we can find SOCIB ocenaographic buoys by specifiying its platfrom_type when queriying the /data-sources/ ENDPOINT as follows: 

```python
#several PARAM examples
end_point = '/data-sources/'
PARAMS = ['platform_type']
PARAMS_values = ['Oceanographic Buoy']

#concatenation of PARAMS
param_concatenation = params_concatenation(PARAMS,PARAMS_values)

#request with more than one PARAM
request_surf = requests.get('%s%s?%s' % (api_url, end_point,param_concatenation), headers=headers)
response_surf = json.loads(request_surf.text)

print('Requested url: '+('%s%s?%s' % (api_url, end_point,param_concatenation)).replace (" ","%20"))
print(' ')
print('SOCIB has performed so far a total of %s deployments involving:' %(response_surf['count']))
for param, value in zip(PARAMS,PARAMS_values):
    print('   '+value +' as '+ param)
```

For Oceanographic buoys, there is a relative stability in the deployments (enssemble platform&instrument) structure: there is a continuity regarding its foundation (platform) and a periodic change of the intruments associated to it. For example, le's see the foundations of the above active deployments: 

```python
#getting the years with Oceanographic Buoys
a = '%s%s?%s' % (api_url, end_point,param_concatenation) 
ids = []
platforms = []
values = []
while a != None:
    #requesting every deployment/data source to check its initial & end datetime
    request_prof = requests.get(a, headers=headers)
    response_prof = json.loads(request_prof.text)
    for drifter in response_prof['results']:
        if drifter['id'] not in ids:
            ids.append(drifter['id'])
            if drifter['platform']['name'] not in platforms:
                platforms.append(drifter['platform']['name'])
                values.append([drifter['id']])
            else:
                values[platforms.index(drifter['platform']['name'])].append(drifter['id'])            
    a = response_prof['next']

for val in values:
    values[values.index(val)] = [len(val),', '.join(val)]
df = pd.DataFrame(values, index=platforms, columns=['times_used','deployments_ids'])
df.style 
```

Let's have a quick look to the deployments associated to one of these foundations:

```python
#getting the years with Oceanographic Buoys
#several PARAM examples
end_point = '/data-sources/'
PARAMS = ['platform_type','status','platform_name']
PARAMS_values = ['Oceanographic Buoy','active','Buoy_CanalDeIbiza']

#concatenation of PARAMS
param_concatenation = params_concatenation(PARAMS,PARAMS_values)
a = '%s%s?%s' % (api_url, end_point,param_concatenation) 
ids = []
all_val = []
while a != None:
    #requesting every deployment/data source to check its initial & end datetime
    request_prof = requests.get(a, headers=headers)
    response_prof = json.loads(request_prof.text)
    for drifter in response_prof['results']:
        if drifter['id'] not in ids:
            ids.append(drifter['id'])
            values = []
            for key in drifter.keys():          
                if key in ['platform','instrument','variables','initial_datetime','end_datetime']:
                    if key == 'platform':
                        values.append(drifter[key]['name'])
                    elif key == 'instrument':
                        values.append(drifter[key]['name']+' ('+drifter[key]['type']['name']+')')
                    else: 
                        values.append(drifter[key])
            all_val.append(values) 
        else:
            continue
    a = response_prof['next']
df = pd.DataFrame(all_val, index=ids, columns=['instrument','variables','platform','initial_datetime','end_datetime']).transpose()
df.style
```

Then, you can work with the data as you please:

```python
end_point = '/data-sources/b9d25cfc3c/data'
PARAMS = ['standard_variable','initial_datetime', 'end_datetime']
PARAMS_values = ['sea_surface_wave_significant_height','2017-06-22T18:00:00','2017-12-07T15:28:00' ]
#concatenation of PARAMS
param_concatenation = params_concatenation(PARAMS,PARAMS_values)
a = '%s%s?%s' % (api_url, end_point,param_concatenation) 
request_prof = requests.get(a, headers=headers)
response_prof = json.loads(request_prof.text)
for key in response_prof.keys():
    print('- With processing level %s:'%(key))
    for feature in response_prof[key]['features']:
        print('Standard variable %s expressed as:' %(PARAMS_values[0]))
        time = feature['coordinates']['time']['data']   
        time = np.array(time, dtype=np.str) #dealing with None values
        for item in feature['variables'][PARAMS_values[0]].keys():
            data = feature['variables'][PARAMS_values[0]][item]['data']
            data = np.array(data, dtype=np.float)#dealing with None values
            
            time2 = [datetime.datetime.strptime(x, '%Y-%m-%dT%H:%M:%S') for x in time]
            
            fig = plt.figure(figsize=(15, 8))
            ax = fig.add_subplot(111)
            plt.title(item + '('+feature['variables'][PARAMS_values[0]][item]['units']+')')
            plt.plot(time2, data, linewidth=2, color="teal")
            plt.xlabel('Year')
            fig.autofmt_xdate()
            plt.grid()
            plt.show()
```
