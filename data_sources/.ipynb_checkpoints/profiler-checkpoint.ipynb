<H3> AUXILIARY FUNCTIONS </H3>

```python
def draw_screen_poly( lats, lons, m):
    """
    Add an rectangle defined by lats, lons to the current m figure 
    """
    x, y = m( lons, lats )
    xy = zip(x,y)
    poly = Polygon( xy, facecolor='white', alpha=0.4, linewidth=5,zorder=3)
    plt.gca().add_patch(poly)
```

```python
def params_concatenation(params, values):
    """
    Concatenates the params and their values in an unique string to later be attached to the generic API ENDPOINTS
    """
    params_concatenation = ''
    for param, value in zip(params,values):
        if params.index(param) != len(PARAMS)-1:
            params_concatenation = params_concatenation + param+'='+value+'&'
        else:
            params_concatenation = params_concatenation + param+'='+value
    return params_concatenation;
```

```python
def highlight_max(data, color='yellow'):
    '''
    highlight the values ove the mean in a Series or DataFrame
    '''
    attr = 'background-color: {}'.format(color)
    if data.ndim == 1:  # Series from .apply(axis=0) or axis=1
        is_max = data > data.mean()
        return [attr if v else '' for v in is_max]
```

<h3> DATA-SOURCES: SURFACE AND PROFILER DRIFTERS</h3>

When looking for specific platforms it is handy to remember the PLATFORM-TYPE ENDPOINT (see<a href="https://github.com/pazrg/SOCIB_API/blob/master/tips/quick_start.ipynb" target="_blank"> Quick start</a>. According to this, there are at least 2 types of drifters available at SOCIB: those operating exclusively in the upper surface of the water column (Surface drifter
), and those able to dive further in it. Disregarding this, let's have an idea about the number of each type available at SOCIB:

```python
#several PARAM examples
end_point = '/data-sources/'
PARAMS = ['platform_type','status']
PARAMS_values = ['Surface drifter','completed']

#concatenation of PARAMS
param_concatenation = params_concatenation(PARAMS,PARAMS_values)

#request with more than one PARAM
request_surf = requests.get('%s%s?%s' % (api_url, end_point,param_concatenation), headers=headers)
response_surf = json.loads(request_surf.text)

print('Requested url: '+('%s%s?%s' % (api_url, end_point,param_concatenation)).replace (" ","%20"))
print(' ')
print('SOCIB has performed so far a total of %s deployments involving:' %(response_surf['count']))
for param, value in zip(PARAMS,PARAMS_values):
    print('   '+value +' as '+ param)
```

```python
#several PARAM examples
end_point = '/data-sources/'
PARAMS = ['platform_type','status']
PARAMS_values = ['Profiler drifter','completed']

#concatenation of PARAMS
param_concatenation = params_concatenation(PARAMS,PARAMS_values)

#request with more than one PARAM
request_prof = requests.get('%s%s?%s' % (api_url, end_point,param_concatenation), headers=headers)
response_prof = json.loads(request_prof.text)

print('Requested url: '+('%s%s?%s' % (api_url, end_point,param_concatenation)).replace (" ","%20"))
print(' ')
print('SOCIB has performed so far a total of %s deployments involving:' %(response_prof['count']))
for param, value in zip(PARAMS,PARAMS_values):
    print('   '+value +' as '+ param)
```

Historically, SOCIB has performed more Surface drifters deloyments than Profiler drifter deployments. <br> As profiler drifters not only drift with surface currents (like Surface drifters) but also provides a vertical profile in the water column, we are going to focuse on those next section.

<h3> Profiler drifters </h3>

Lets get a quick view of the Profiler drifter abundance by year and month of SOCIB Profiler drifters:

```python
#several PARAM examples
end_point = '/data-sources/'
PARAMS = ['platform_type','status']
PARAMS_values = ['Profiler drifter','completed']

#concatenation of PARAMS
param_concatenation = params_concatenation(PARAMS,PARAMS_values)

#getting the years with Profilers drifters
a = '%s%s?%s' % (api_url, end_point,param_concatenation) 
years = []
while a != None:
    #requesting every deployment/data source to check its initial & end datetime
    request_prof = requests.get(a, headers=headers)
    response_prof = json.loads(request_prof.text)
    for drifter in response_prof['results']:
        ini_year = int(drifter['initial_datetime'].split('T')[0].split('-')[0])
        end_year = int(drifter['end_datetime'].split('T')[0].split('-')[0])
        if ini_year not in years:
            years.append(ini_year)
        if end_year not in years:
            years.append(end_year)
    a = response_prof['next']

#getting the number of Profiler drifters per year and month with the API (ENDPOINT + PARAMS)
serie_prf = []
for year in np.sort(years):
    yearly_serie = []
    for month in range(1,13):
        end_point = '/data-sources/'
        PARAMS = ['platform_type','status','initial_datetime','end_datetime']
        PARAMS_values = ['Profiler drifter','completed',str(year)+'-'+str(month).zfill(2) +'-01',str(year)+'-'+str(month).zfill(2) +'-'+str(monthrange(year,month)[1]).zfill(2)]

        #concatenation of PARAMS
        param_concatenation = params_concatenation(PARAMS,PARAMS_values)

        #request with more than one PARAM
        request_prof = requests.get('%s%s?%s' % (api_url, end_point,param_concatenation), headers=headers)
        response_prof = json.loads(request_prof.text)
        yearly_serie.append(response_prof["count"])
    serie_prf.append(yearly_serie) 

print('Quick view of the Profiler drifter abundance by year and month')
df = pd.DataFrame(serie_prf, index=np.sort(years), columns=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
df.style
df.style.apply(highlight_max)
```

In terms of time intervals:

```python
#several PARAM examples
end_point = '/data-sources/'
PARAMS = ['platform_type','status']
PARAMS_values = ['Profiler drifter','completed']

#concatenation of PARAMS
param_concatenation = params_concatenation(PARAMS,PARAMS_values)

a = '%s%s?%s' % (api_url, end_point,param_concatenation) 
time_series = []
while a != None:
    #requesting the page where the deployment is
    request_prof = requests.get(a, headers=headers)
    response_prof = json.loads(request_prof.text)
    for drifter in response_prof['results']:
        drifter_id = drifter['id']
        request_detailed_drifter = requests.get('%s%s%s/data/' % (api_url, end_point,drifter_id), headers=headers)
        response_detailed_drifter = json.loads(request_detailed_drifter.text)
        
        for feature in response_detailed_drifter['L1']['features']:
            if feature['feature_type'] == 'trajectory':
                time_series.append([drifter_id,feature['coordinates']['time']['data']])
    a = response_prof['next']

ids = []
idx = pd.Index(time_series[0][1])
for time in time_series: 
    idx = idx.union(pd.Index(time[1]))
    ids.append(time[0])
df = pd.DataFrame(index=idx)

for time in time_series: 
    df[time[0]] = df.index.to_series().apply(lambda x: time_series.index(time) if x >= min(time[1]) and x <= max(time[1]) else np.NaN)   
# plot
p = df.plot(legend=False, figsize=(20,10), fontsize=20, linewidth = 5)
p.grid(True)

p.set_ylim(p.get_ylim()[0]-1,p.get_ylim()[1]+1)
p.set_yticklabels(['']+ids+[''])
plt.yticks(np.arange(-1, len(['']+ids+[''])-1, 1.0))

for time in time_series:
    p.text(idx.tolist().index(time[1][0]),ids.index(time[0])+0.15, time[1][0],fontsize=15)
    p.text(idx.tolist().index(time[1][len(time[1])-1])+0.15,ids.index(time[0])-0.15, time[1][len(time[1])-1], fontsize=15)
print('drifter identifier (id) vs time window')
```

Checking the trajectory points of the above deployments: 

```python
fig = plt.figure(figsize=(20,16))
m = Basemap(projection='merc', llcrnrlat=lat_min-5, urcrnrlat=lat_max+5,llcrnrlon=lon_min-5, urcrnrlon=lon_max+5, resolution='i')
m.drawcoastlines(linewidth=0.5, zorder=3)
m.fillcontinents(zorder=2)
m.drawparallels(np.arange(-90.,91.,2.), labels=[1,0,0,0], linewidth=0.5, zorder=1, fontsize=20)
m.drawmeridians(np.arange(-180.,181.,3.), labels=[0,0,1,0], linewidth=0.5, zorder=1, fontsize=20)
lats = [ lat_max, lat_min, lat_min, lat_max ] 
lons = [ lon_min, lon_min, lon_max, lon_max] 
draw_screen_poly( lats, lons, m )

#several PARAM examples
end_point = '/data-sources/'
PARAMS = ['platform_type','status','bbox']
PARAMS_values = ['Profiler drifter','completed']

#concatenation of PARAMS
param_concatenation = params_concatenation(PARAMS,PARAMS_values)

#BoundingBox
bounding_box = shapely.geometry.box(lon_min,lat_min,lon_max, lat_max)

ids = []
a = '%s%s?%s' % (api_url, end_point,param_concatenation) 
while a != None:
    request_prof = requests.get(a, headers=headers)
    response_prof = json.loads(request_prof.text)
    for drifter in response_prof['results']:
        drifter_id = drifter['id']
        #details of each profiler drifter retourned
        request_detailed_drifter = requests.get('%s%s%s/data/' % (api_url, end_point,drifter_id), headers=headers)
        response_detailed_drifter = json.loads(request_detailed_drifter.text)
        for feature in response_detailed_drifter['L1']['features']:
            if feature['feature_type'] == 'trajectory':
                latitude = feature['coordinates']['latitude']['data']
                longitude = feature['coordinates']['longitude']['data']
                latitude = np.array(latitude, dtype=np.float)#dealing with None values
                longitude = np.array(longitude, dtype=np.float)#dealing with None values                
                #cheking if any coordinate is within the boundingbox
                checker = []
                for lon,lat in zip(longitude, latitude):
                    xy_point = Point(lon,lat)
                    checker.append(bounding_box.contains(xy_point))
                #Plot the points on the map if at least one of the points
                if True in checker: 
                    x, y = m(longitude, latitude)
                    m.plot(x,y,marker='o', linewidth=3, label=drifter_id)
                    ids.append(drifter_id)

    a = response_prof['next']
lg = plt.legend()
lg.get_frame().set_facecolor('white')
plt.show()
print('Just %s Profiler drifter have been around that area: %s'%(len(ids), ','.join(ids)))
```

From coordinates is easy to identify which variables has been measured in the ocean upper layer (no depth reference) and those reporting measured in the water column (depth reference). Let's have a look to both types in one of the above Profiler drifters:

```python
request_detailed_drifter = requests.get('%s/data-sources/%s/data/' % (api_url,'5961a15c40'), headers=headers)
response_detailed_drifter = json.loads(request_detailed_drifter.text)
```

- PROFILE DATA

```python
feature_type = 'trajectoryProfileIrregular'
product_level = 'L1'

for element in response_detailed_drifter[product_level]['features']:
    if element['feature_type'] == feature_type:
        #getting time and depth 
        time = feature['coordinates']['time']['data']
        time = np.array(time, dtype=np.str) #dealing with None values
        depth = feature['coordinates']['depth']['data']
        depth = np.array(depth, dtype=np.float)#dealing with None values

        for variable in element['variables'].keys():
            for param in element['variables'][variable].keys():
                print(param+ ' in '+element['variables'][variable][param]['units']+' ('+variable+')')
                fig = plt.figure(figsize=(10,8))
                
                #color_bar
                cmap = plt.cm.Spectral_r
                maxima = 0
                minima = 1000000000
                for ntime in range(0, len(time)):
                    if max(element['variables'][variable][param]['data'][ntime]) > maxima :
                        maxima = max(element['variables'][variable][param]['data'][ntime]) 
                    if min(element['variables'][variable][param]['data'][ntime]) < minima :
                        minima = min(element['variables'][variable][param]['data'][ntime])        
                norm = colors.Normalize(vmin=minima, vmax=maxima)
                
                #drawing profile
                for ntime in range(0, len(time)):
                    plt.scatter(time.tolist().index(time[ntime])*np.ones_like(depth[ntime]), depth[ntime], s=20, c=element['variables'][variable][param]['data'][ntime], edgecolor='None', cmap=cmap, norm=norm)
                plt.ylim(0.0, 1200)
                plt.gca().invert_yaxis()
                plt.colorbar(extend='both')
                fig.autofmt_xdate()
                plt.ylabel("%s (%s)" % ('depth',feature['coordinates']['depth']['units']))
                plt.xlabel("%s" % ('time'))
                plt.xticks(range(0, len(time)), time)
                plt.show()
```
