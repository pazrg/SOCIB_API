<h3>HANDY PYTHON PACKAGES</h3>

```python
import requests
import json
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap
from json2html import *
from IPython.display import HTML
%matplotlib inline
```

<h3>PREREQUISITES</h3>

```python
api_key = '' #write you SOCIB api_key
api_url = 'http://api.socib.es'
headers = {
    'accept': 'application/vnd.socib+json',
    'apikey': api_key,
}
```

<h3> GETTING SOCIB FULL-LIST OF PLATFOMRS</h3>

```python
platforms = []
page = "http://api.socib.es/data-platforms/?page=1"
while page != None:
    request = requests.get('%s' % (page), headers=headers)
    response = json.loads(request.text)
    print('Requested url: '+'%s' % (page))
    for platform in response['results']:
        platforms.append(platform)
    page = response['next']
HTML(json2html.convert(json=platforms))
```

<h3> GETTING SOCIB FULL-LIST OF A CERTAIN TYPE OF PLATFOMRS</h3>

```python
platforms = []
page = "http://api.socib.es/data-platforms/?platform_type=Surface drifter&page=1"
while page != None:
    request = requests.get('%s' % (page), headers=headers)
    response = json.loads(request.text)
    print('Requested url: '+'%s' % (page))
    for platform in response['results']:
        platforms.append(platform)
    page = response['next']
HTML(json2html.convert(json=platforms))
```
